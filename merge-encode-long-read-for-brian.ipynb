```python
def get_alignment_library_map(experiments):
    alignment_to_library = {}
    for experiment_accession in set(experiments):
        experiment = server.get_experiment(experiment_accession)
        for replicate in experiment.replicates:
            for f in replicate.files:
                if f.output_type == "alignments":
                    alignment_to_library[f.accession] = replicate.library["accession"]
    return alignment_to_library
```

```python
fairlie_human_metadata = pandas.read_csv("https://raw.githubusercontent.com/fairliereese/paper_rnawg/master/proc_revisions/data/human/lr/lr_human_library_data_summary.tsv", sep="\t")
print(fairlie_human_metadata.shape)

fairlie_human_metadata.head()
```

```python
human_alignment_library = get_alignment_library_map(fairlie_human_metadata["ENCODE_experiment_id"])
```

```python
human_alignment_library
```

```python
fairlie_human_metadata["library"] = fairlie_human_metadata["ENCODE_alignments_id"].apply(lambda x: human_alignment_library.get(x))
fairlie_human_metadata.head()
```

```python
human_library = fairlie_human_metadata.pop("library")
fairlie_human_metadata.insert(1, "library", human_library)
fairlie_human_metadata = fairlie_human_metadata.drop_duplicates(ignore_index=True).sort_values("dataset")
fairlie_human_metadata
```

```python
pacbio_human = pandas.read_excel("ENCODE PacBio Libraries.xlsx", sheet_name="Human datasets")
pacbio_human
```

```python
fairlie_human_metadata = fairlie_human_metadata.merge(pacbio_human, left_on="library", right_on="ENCODE library accession", how="left")
```
