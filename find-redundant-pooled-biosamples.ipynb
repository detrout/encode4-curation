```python
import pandas
import os
import sys
from pathlib import Path
```

```python
pandas.options.display.max_rows = 100
```

```python
mouse_biosamples = Path("~/woldlab/ENCODE/mouse-biosamples").expanduser()
```

```python
metadata_translation_filename = mouse_biosamples / "Wold mouse tissue metadata translation table.xlsx"
translation = pandas.read_excel(
    metadata_translation_filename, 
    skiprows=1,
    dtype={"Unnamed: 15": str}
)
columns = list(translation.columns)
columns[-1] = "comment"
translation.columns = columns
```

```python
accessions = []
sent = []
for accession in translation["ENCBS"]:
    if pandas.isnull(accession):
        accessions.append(accessions[-1])
        sent.append(sent[-1])
    else:
        accession = accession.split()
        accessions.append(accession[0])
        if len(accession) > 1 and accession[3] == "Brian)":
            sent.append(True)
        else:
            sent.append(False)

#print(accessions)
#print(sent)
```

```python
translation["ENCBS"] = accessions
translation["sent"] = sent
```

```python
pooled = pandas.read_excel(mouse_biosamples / "BioSample_metadata_for_pooled_samples_Jan12_2022.xlsx")
pooled
```

```python
pooled_map = {}
for i, row in pooled.iterrows():
    for pooled_from in row["pooled_from:array"].split(","):
        pooled_map[pooled_from] = row["accession"]
```

```python
translation_map = {}
for i, row in translation.iterrows():
    pooled_from = row["Pooled_from"]
    if not pandas.isnull(pooled_from):
        translation_map[pooled_from] = row["ENCBS"]
```

```python
translation_map
```

```python
replaced_by = {}
for pooled_from in translation_map:   
    replaced_by.setdefault(translation_map[pooled_from], set()).add(pooled_map[pooled_from])
    
replaced_by
```

```python
for theirs in replaced_by:
    ours = list(replaced_by[theirs])[0]
    print(theirs, "replaced", ours)
```

```python
for theirs in replaced_by:
    ours = list(replaced_by[theirs])[0]
    print("https://www.encodeproject.org/biosamples/{}/".format(theirs), "https://www.encodeproject.org/biosamples/{}/".format(ours))
```
