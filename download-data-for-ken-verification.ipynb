```python
def get_score(experiment, name=None, columns=["TPM"]):
    for replicate in experiment.replicates:
        for f in replicate.files:
            if f["output_type"] == "gene quantifications":
                url = server.prepare_url(f["href"])
                rsem = pandas.read_csv(url, sep="\t", index_col="gene_id")
                for column in columns:
                    series = rsem[column]
                    if name is None:
                        donor = replicate["library"]["biosample"]["donor"]
                        rush_id = experiment["accession"]
                        for candidate in donor["aliases"]:
                            if candidate.startswith("john-stamatoyannopoulos:E"):
                                rush_id = candidate.replace("john-stamatoyannopoulos:", "")
                                
                        series.name = "{}_{}".format(
                            rush_id, 
                            replicate.library["accession"], 
                        )
                    else:
                        series.name = name
                    yield (column, series)
       
```

```python
for quant, matrix in list(get_score(exp, columns=["TPM", "expected_count"])):
    print(quant, matrix.name, matrix.shape)
```

```python
scores = {}

for rush_id, row in merged.iterrows():
    experiment = server.get_experiment(row["accession"])
    for quantification, matrix in get_score(experiment, columns=["TPM", "expected_count"]):
        scores.setdefault(quantification, []).append(matrix)
```

```python
tpms = pandas.DataFrame(scores["TPM"]).T
expected_count = pandas.DataFrame(scores["expected_count"]).T
```
