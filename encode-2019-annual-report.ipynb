```python
progress = []
for summary in current['@graph']:
    obj = server.get_json(summary['@id'])
    progress.append({
        'accession': obj['accession'],
        'lab': obj['lab']['@id'],
        'assay_title': obj['assay_title'],
        'description': obj['description'],
        'biosample': obj['biosample_summary'],
        'date_created': obj['date_created'].split('T')[0],
        'date_submitted': obj.get('date_submitted', None),
        'date_released': obj.get('date_released', None),
        'files_available': len(obj.get('files', [])),
        'status': obj['status'],
        'audit_error': [ x['category'] for x in summary.get('audit', {}).get('ERROR', [])]
    })
progress = pandas.DataFrame(progress)
progress.set_index('accession', inplace=True)
```

```python
progress[progress['status'] != 'archived'].sort_values(['date_released', 'date_created'])
```

```python
progress[progress['status'] != 'archived'].sort_values(['date_released', 'accession']).to_excel('encode4-progress.xlsx')
```
