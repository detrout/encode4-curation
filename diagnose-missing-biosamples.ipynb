```python
for i, row in mouse_cortex_library.iterrows():
    library = server.get_json("/library/{}/".format(row['accession']))
    biosample = library.get("biosample")
    if biosample is not None:
        biosample = biosample['accession']
    if biosample != row['biosample']:
        print(row['accession'], biosample, row['biosample'])
        #print(server.patch_json(library["@id"], {"biosample": row['biosample']}))
        #print()
    
```

```python
for i, row in mouse_cortex_experiments.iterrows():
    print(server.prepare_url('/experiment/{}/'.format(row['accession'])))
```

```python
mouse_cortex_replicates = pandas.read_excel(sheet_file, sheet_name="Replicate")
report = ["<table><tr><td>Exp Desc</td><td>Bio Desc</td><td>Age</td><td>Age Units</td></tr>"]
for i, row in mouse_cortex_replicates.iterrows():
    replicate = server.get_json(row['uuid'])
    experiment = replicate['experiment']
    biosample_accession = replicate['library']['biosample']['accession']
    biosample = server.get_json('/biosample/{}/'.format(biosample_accession))
    report.append("<tr><td>{}</td><td>{}</td><td>{}</td><td>{}</td></tr>".format(
        experiment['description'], 
        biosample['description'], 
        biosample['model_organism_age'], 
        biosample['model_organism_age_units']
    ))
report.append("</table>")
HTML("\n".join(report))
```
